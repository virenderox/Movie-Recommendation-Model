```python
text = ["London Paris London","Paris London Paris"] 
```

```python
# import the sklearn to count the frequecy in text
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# CountVectorizer is class having method to count frequeny
cv =  CountVectorizer()
```

```python
# using methods fit_transforms to find the frequency 
count_matrix = cv.fit_transform(text)
```

```python
count_matrix.toarray()
```

```python
#import the sklearn to cosine_similarity between two vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(count_matrix)
similarity_score
```

```python
# Alogrithm for recomndation system

#step 1: Read csv file
#step 2: selection Features
#step 3: Create the column in df which combine all selection features
#step 4: create count_matrix for this new combined columns
#step 3: Compute the cosine_similarity based upon the coulum matrix

```

```python
import pandas as pd
import numpy as np
```

```python
# step 1
data = pd.read_csv("C:\\Users\\Virender Pal Singh\\Desktop\\Document\\machine_learing\\movie_dataset.csv")
```

```python
data.head()
```

```python
# step 2 
# selection featurs
features = ['keywords', 'cast', 'genres', 'director']
```

```python
# step 3

for feature in features:
    data[feature] = data[feature].fillna('')
def combine_features(row):
    try:
        return row['keywords'] + " " + row['cast'] + " " + row["genres"] + " " + row["director"]
    except:
        print("Error:", row)
data["combine_features"] = data.apply(combine_features, axis=1)
print("Combined Features:", data["combine_features"].head())
```

```python
# step 4
cv = CountVectorizer()
count_matrix = cv.fit_transform(data["combine_features"])
count_matrix.toarray()
```

```python
# step 5

cosine_sim = cosine_similarity(count_matrix)
cosine_sim
```

```python
# helper function 

def get_title_from_index(index):
    return data[data.index == index]["title"].values[0]
```

```python
def get_index_from_title(title):
    return data[data.title == title]["index"].values[0]
```

```python
movie_user_like = "The Dark Knight Rises"
```

```python
# step 6
# Get index of this movie from its title
movie_index =  get_index_from_title(movie_user_like)
movie_index
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
similar_movies
```

```python
# step 7
# sorted the movie in descending order
sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1],reverse = True)
sorted_similar_movies
```

```python
# step 8 :Print titles of first 50 movies
i = 0
for movie in sorted_similar_movies:
    print(get_title_from_index(movie[0]))
    i = i + 1
    if i > 50:
        break
        
```
